```python
#20. 데이터를 로드하라

dc = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv")
dc.head(3)
```

```python
#21. quantity 컬럼 값이 3인 데이터를 추출하여 첫 5행을 출력하라

dc[dc.quantity == 3].head(5)
```

```python
#22. quantity 컬럼값이 3인 데이터를 추출하여 index를 0부텆 정렬하고 첫 5행을 출력

dc[dc.quantity == 3].sort_index().head(5)
```

```python
#23. quantity, item_price 두개의 컬럼으로 구성된 새로운 데이터 프레임을 정의하라

dc[['quantity','item_price']]
```

```python
#24. item_price 컬럼의 달러표시 문자를 제거하고 float 타입으로 저장하여 new_price 컬럼에 저장하라

dc['new_price'] = dc.item_price.str.removeprefix("$").astype(float)
dc.info()
```

```python
#25. new_price 컬럼이 5이하의 값을 가지는 프레임을 추출하고, 전체 개수를 구하라

mast = dc.new_price <= 5
len(dc[mast])
```

```python
#26. item_name 명이 Chicken Salad bowl 인 데이터 프레임을 추출하라고 index값을 초기화 하여라

mask = dc.item_name == 'Chicken Salad Bowl'
dc[mask].reset_index(drop=True)
```

```python
#27. new_price 값이 9이하이고 item_name 값이 Chicken Salad Bowl 인 데이터 프레임 추출
```

```python
mask = (dc.new_price <= 9) & (dc.item_name == 'Chicken Salad Bowl')
dc[mask].head(3)
```

```python
#28. df의 new_price 컬럼값에 따라 오름차순으로 정리하고 index를 초기화 해라
```

```python
dc.new_price.sort_values().reset_index(drop=True)
```

```python
#29. dc 의 item_name 컬럼 값중 chips를 포함하는 경우 데이터를 출력하라

mask = dc.item_name.str.contains("Chips")
dc[mask]
```

```python
#30. df 의 짝수번째 컬럼만을 포함하는 데이터프레임을 출력하라

dc.iloc[:, ::2]
```

```python
# 31. dc의 new_price 컬럼 값에 따라 내림차순으로 정리하고 index를 초기화 하라

# dc.new_price.sort_values(ascending=False).reset_index()
dc.sort_values(by='new_price', ascending=False).reset_index()
```

```python
#32. df 의 item_name 컬럼 값이 streak salad or Bowl인 데이터를 인덱싱하라

mask = (dc.item_name == 'Steak Salad') | (dc.item_name == 'Bowl')
dc[mask]
```

```python
dc[dc.item_name.isin(['Steak Salad', 'Bowl'])]
```

```python
#33. df의 item_name 컬럼 값이 streak salad or bowl인 데이터를 데이터 프레임화한후 
# item_name을 기준으로 중복행이 있으면 제거하되, 첫번째 케이스만 남겨라.

dc[dc.item_name.isin(['Steak Salad', 'Bowl'])].drop_duplicates(subset='item_name', keep='first')
```

```python
#34.dc의 item_name 컬럼 값이 streak salad or bowl인 데이터를 데이터 프레임화한후 
# item_name을 기준으로 중복행이 있으면 제거하되, 마지막 케이스만 남겨라.

dc[dc.item_name.isin(['Steak Salad', 'Bowl'])].drop_duplicates(subset='item_name', keep='last')
```

```python
#35. df 데이터 중 new_price 값이 new_price 값의 평균값 이상을 가지는 데이터를 인덱싱하라

dc[dc.new_price >= dc.new_price.mean()].head(3)
```

```python
#36. df 데이터 중 item_name 값이 lzze 데이터를 Fizzy Lizzy로 수정하라

dc.item_name.str.replace("Izze", "Fizzy Lizzy")
```

```python
#37. dc 데이터 중 choice_description 값이 NaN인 데이터의 개수는?

dc.choice_description.isnull().sum()
```

```python
#38. df 데이터 중 choice_description 값이 NaN인 데이터를 NoData 값으로 대체하라(loc 이용)

dc[dc.choice_description.isnull()]['choice_description'].str.replace("NaN", "NoData")
```
